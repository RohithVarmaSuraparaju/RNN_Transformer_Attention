# Q1: Character-Level RNN Language Model (Example Outputs)

```
# Toy dataset
text = "hello help hello world hello rnn"

chars = sorted(list(set(text)))
print("Vocabulary:", chars)

# Example training info
print("Hidden size:",128)
print("Sequence length:",50)
print("Epochs:",5)
```

```
# Example training loss
print("Epoch 1 Loss: 2.52")
print("Epoch 2 Loss: 2.31")
print("Epoch 3 Loss: 2.10")
print("Epoch 4 Loss: 1.92")
print("Epoch 5 Loss: 1.75")
```
